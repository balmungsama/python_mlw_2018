# Dimension reduction: PCA vs ICA

![pca+ica](./images/pca_vs_ica.png)


Both [PCA]() and [ICA]() are statistical transformations, usually used to reduce the size of the feature space.

* **PCA:** information from second order statistics
    - data-centered ideal space for the coordinate representations
* **ICA:** information that goes up to high order statistics

## PCA
* Finds principal components of the dataset
    - high variance component explains more variance in the signal
* Each succeeding component points in direction that <span style="color:red">**explains most variance**</span>.
* Transforms data into new subspace
* First axis corresponds to first principal component (explains greatest amount of variance in the data)
    - every axis will be a linear combination of the original axes
* like regularization, but PCA is a priori reduction of dimension to improve your model fit
* components are weighted by the relative contribution with beta coefficients
    - beta coefficients are a linear combination of the original dataset

## ICA
Assume there exist independent signals: $ùëÜ =[ùë†_1ùë° , ùë†_2ùë° , ‚Ä¶ , ùë†_ùëÅùë°]$ that we can recover from our data.

    - independent: making an independent contribution to the outcome
Non-guassian signals 
However, what we actually observe is a linear combination of the original signals with some mixing matrix, such that what we see is the mixing applied to the original independent data: $ùëå(ùë°) = ùê¥*ùëÜ(ùë°)$.

We know neither what the mixing matrix $A$ nor the original source signals $S$ are.

The goal of ICA is to <span style="color:red"> **recover the original signals** </span>, $S(t)$ from $Y(t)$.
    - removing artifacts

## PCA vs ICA differences

* PCA removes correlations, but <span style="color:red">not</span> higher order dependence
* ICA removes correlations and higher order dependence
* PCA: some components are more important than others (recall eigenvalues)
* ICA: all components are equally important, no way to "rank" components in terms of importance
* PCA: vectors are orthogonal (recall eigenvectors of covariance matrix)
* ICA: vectors are not orthogonal (instead, dimensions are maximally non-gaussian)

### Biggest difference is: PCA looks for uncorrelated dimensions in your data, ICA looks for independent dimensions in your data (independence here means maximally non-gaussian). So, if you expect your source signals to be non-gaussian, then ICA may be a good method to use, otherwise, stear clear!

# ICA vs PCA on 2D point clouds


This first example illustrates visually in the feature space a comparison by results using two different component analysis techniques.

`ICA` vs `PCA`.

Representing ICA in the feature space gives the view of 'geometric ICA': ICA is an algorithm that finds directions in the feature space **corresponding to projections with high non-Gaussianity**. These directions need not be orthogonal in the original feature space, but they are orthogonal in the whitened feature space, in which all directions correspond to the same variance.

**PCA, on the other hand, finds orthogonal directions in the raw feature space that correspond to directions accounting for maximum variance.**

Here we simulate independent sources using a highly non-Gaussian process, 2 student T-distributions with a low number of degrees of freedom (top left figure). We mix them to create observations (top right figure).

In this raw observation space, directions identified by PCA are represented by orange vectors. We represent the signal in the PCA space, after whitening by the variance corresponding to the PCA vectors (lower left). Running ICA corresponds to finding a rotation in this space to identify the directions of largest non-Gaussianity (lower right).



```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA, FastICA
```

```python
# #############################################################################
# Generate sample data
rng = np.random.RandomState(42)
S = rng.standard_t(2, size=(20000, 2))
S[:, 0] *= 2.
S.shape
```

```python
# Mix data
A = np.array([[1, 1], [0, 2]])  # Mixing matrix
print(A.T)
X = np.dot(S, A.T)  # Generate observations
```

```python
pca = PCA()
S_pca_ = pca.fit(X).transform(X)

ica = FastICA(random_state=rng)
S_ica_ = ica.fit(X).transform(X)  # Estimate the sources

S_ica_ /= S_ica_.std(axis=0)
```

```python
def plot_samples(S, axis_list=None):
    plt.scatter(S[:, 0], S[:, 1], s=2, marker='o', zorder=10,
                color='steelblue', alpha=0.5)
    if axis_list is not None:
        colors = ['orange', 'red']
        for color, axis in zip(colors, axis_list):
            axis /= axis.std()
            x_axis, y_axis = axis
            # Trick to get legend to work
            plt.plot(0.1 * x_axis, 0.1 * y_axis, linewidth=2, color=color)
            plt.quiver(0, 0, x_axis, y_axis, zorder=11, width=0.01, scale=6,
                       color=color)

    plt.hlines(0, -3, 3)
    plt.vlines(0, -3, 3)
    plt.xlim(-3, 3)
    plt.ylim(-3, 3)
    plt.xlabel('x')
    plt.ylabel('y')
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2, 2, 1)
plot_samples(S / S.std())
plt.title('True Independent Sources')

axis_list = [pca.components_.T, ica.mixing_]
plt.subplot(2, 2, 2)
plot_samples(X / np.std(X), axis_list=axis_list)#whiten component vector projections
legend = plt.legend(['PCA', 'ICA'], loc='upper right')
legend.set_zorder(100)

plt.title('Observations')

plt.subplot(2, 2, 3)
plot_samples(S_pca_ / np.std(S_pca_, axis=0))#whiten the pca data
plt.title('PCA recovered signals')

plt.subplot(2, 2, 4)
plot_samples(S_ica_ / np.std(S_ica_))#whiten ica data
plt.title('ICA recovered signals')

plt.subplots_adjust(0.09, 0.04, 0.94, 0.94, 0.26, 0.36)
plt.show()
```

# ICA vs PCA on [iris dataset](https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.names)

```python
from sklearn.datasets import load_iris
import pandas as pd
import seaborn as sns
```

```python
iris_data = load_iris()
```

```python
iris_data.keys()
```

```python
iris_df = pd.DataFrame(iris_data["data"],columns=iris_data["feature_names"])
```

```python
# center the data 
iris_X = iris_df - iris_df.mean()
```

```python
iris_X.mean()
```

```python
iris_X["target"] = iris_data["target"]
```

## Plot iris data in original feature space

```python
sns.pairplot(iris_X,vars = iris_X.columns[:-1],hue="target");
```

## Plot iris data in PCA and ICA spaces

```python
iris_pca = PCA()
iris_pca_X = iris_pca.fit_transform(iris_X.iloc[:,:-1])

iris_ica = FastICA(random_state=rng,max_iter=1000)
iris_ica_X = iris_ica.fit_transform(iris_X.iloc[:,:-1])
```

```python
iris_pca_X = pd.DataFrame(iris_pca_X,columns=["pca_1","pca_2","pca_3","pca_4"])
iris_pca_X["target"] = iris_data["target"]
iris_ica_X = pd.DataFrame(iris_ica_X,columns=["ica_1","ica_2","ica_3","ica_4"])
iris_ica_X["target"] = iris_data["target"]
```

## PCA space

```python
sns.pairplot(iris_pca_X,vars = iris_pca_X.columns[:-1],hue="target");
```

```python
# variance in the original data captured by each component 
iris_pca.explained_variance_ratio_
```

```python
sns.pairplot(iris_ica_X,
             vars = iris_ica_X.columns[:-1],
             hue="target")
```
